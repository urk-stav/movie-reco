```
# libraries
import json
import wikipediaapi
from copy import deepcopy
from urllib.parse import unquote
```

```
################################################################################
# wiki_movies_infos
# list of films by year by country scraped from wikipedia
################################################################################

with open('/content/drive/My Drive/wiki_scraper/wiki_movie_infos.json', 'r') as f:
  movie_infos = json.load(f)
```

```
n = sum([len([m for _v in v.values() for m in _v]) for v in movie_infos.values()])
n
```

```
movie_plots = {}
```

```
################################################################################
# scrapes text from wikipedia pages for the movies
# timeouts happen again and again, need to fix that
# for now, just re-run this cell
################################################################################

wiki_wiki = wikipediaapi.Wikipedia('en')

_n = 0
for _, v in movie_infos.items():
  for country_year, movies in v.items():
    if country_year not in movie_plots:
      movie_plots[country_year] = {}
    for m in movies:
      if not m['url'].startswith('/w/index.php'):
        url = unquote(m['url'][6:])
        perc_done = f'{round(_n / n * 100, 3)}%'
        if m['url'] not in movie_plots[country_year]:
          print(f"Collecting [{country_year}] {m['url']} [{perc_done} done]...")
          page_py = wiki_wiki.page(url)
          text = page_py.text
          movie_plots[country_year][m['url']] = text
        else:
          print(f"Skipping [{country_year}] {m['url']} !")
      _n += 1
```

```
len([v1 for k, v in movie_plots.items() for k1, v1 in v.items()])
```

```
# write out data
with open('/content/drive/My Drive/wiki_scraper/movie_plots.json', 'w') as f:
  json.dump(movie_plots, f)

files.download('/content/drive/My Drive/wiki_scraper/movie_plots.json')
```
